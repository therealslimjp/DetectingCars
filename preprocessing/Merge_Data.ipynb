```python
import os
import shutil

# function to merge the two datasets with the street masks and the car masks into one dataset
# moves the street masks from the street dataset into the car dataset
def merge_datasets(car_dataset_path, street_dataset_path):
    car_splits = ['train', 'test', 'validation']
    street_splits = ['train', 'test', 'validation']

    # Walk through all folders and subfolders in car dataset
    for car_split in car_splits:
        car_split_path = os.path.join(car_dataset_path, car_split)

        for root_car, dirs_car, _ in os.walk(car_split_path):
            for dir_name in dirs_car:
                car_dir = os.path.join(root_car, dir_name)

                # Walk through all folders and subfolders in street dataset
                for street_split in street_splits:
                    street_split_path = os.path.join(street_dataset_path, street_split)
                    street_dir = os.path.join(street_split_path, dir_name)

                    # check if street directory exists and move all street masks into car directory
                    if os.path.exists(street_dir):
                        street_files = os.listdir(street_dir)
                        for street_file in street_files:
                            if street_file.endswith('.npy'):
                                shutil.move(os.path.join(street_dir, street_file), car_dir)
                                print(f"Moved {street_file} to {car_dir}")
                    else:
                        print(f"Street directory not found for {car_dir}")


```

```python
# merge the two datasets with the street masks and the car masks into one dataset
car_dataset_path = "../preprocessing/train_test_split_car"
street_dataset_path = "../preprocessing/train_test_split"
merge_datasets(car_dataset_path, street_dataset_path)
```

```python
# function to delete all img.png files so save space
def delete_img_png_files(directory):
    # Walk through all folders and subfolders
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file == "img.png":
                file_path = os.path.join(root, file)
                os.remove(file_path)
                print(f"Deleted {file_path}")
```

```python
#delete all img.png files
delete_img_png_files("../preprocessing/train_test_split_car")
```
